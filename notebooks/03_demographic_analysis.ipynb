```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/clean_dataset.csv")
df.head()
```

```python
# Gender
df["Gender_Label"].value_counts().plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Distribution of Gender")
plt.ylabel("Count")
plt.show()

# University
df["University_Label"].value_counts().plot(kind="bar", color="orange", edgecolor="black")
plt.title("Type of University")
plt.ylabel("Count")
plt.show()

# Education Level
df["Education_Label"].value_counts().plot(kind="bar", color="green", edgecolor="black")
plt.title("Level of Education")
plt.ylabel("Count")
plt.show()
```

```python
# Province (Top 10)
df["Province_Label"].value_counts().head(10).plot(kind="barh", color="purple", edgecolor="black")
plt.title("Top 10 Provinces of Respondents")
plt.xlabel("Count")
plt.show()

# Fields of Study
df["Field_Label"].value_counts().plot(kind="bar", color="teal", edgecolor="black")
plt.title("Distribution of Fields of Study")
plt.ylabel("Count")
plt.show()
```

```python
df["AI_Label"].value_counts().plot(kind="bar", color="red", edgecolor="black")
plt.title("Types of AI Tools Used")
plt.ylabel("Count")
plt.show()
```

```python
# Gender Ã— AI Usage
ct_gender_ai = pd.crosstab(df["Gender_Label"], df["AI_Label"])
print(ct_gender_ai)

sns.heatmap(ct_gender_ai, annot=True, cmap="Blues", fmt="d")
plt.title("Cross-tab: Gender vs AI Usage")
plt.show()

# Province Ã— University
ct_province_uni = pd.crosstab(df["Province_Label"], df["University_Label"])
ct_province_uni.head(10) 
```
